```matlab
axis([-1 1 -1 1])
axis square
hold on
% that is how a 2D rotation matrix can be defined
trplot2(eye(2), 'frame', '0', 'color', 'black') % origin frame
trplot2(rot2(pi/4), 'frame', '1', 'color', 'r') % counter-clockwise rotation for 45 degrees

% inverse rotation is equivalent to the transpose matrix
trplot2(rot2(pi/3)', 'frame', '2', 'color', 'b') % clockwise rotation for 60 degrees
```

```matlab
% create a 2D homogeneous transformation matrices (translation along X and Y axes + rotation)
T1 = SE2(1, 2, 30, 'deg')
T2 = SE2(2, 1, 0, 'deg')
T3 = T1 * T2
T4 = T2 * T1
```

```matlab
axis([0 5 0 5])
axis square
hold on

trplot2(T1, 'frame', '1', 'color', 'b')
trplot2(T2, 'frame', '2', 'color', 'g')
trplot2(T3, 'frame', '3', 'color', 'r')
trplot2(T4, 'frame', '4', 'color', 'c')

% create a point in the reference frame
P = [3; 2]

plot_point(P, '*', 'label', 'P')

% determine the coordinate of a point P w.r.t to coordinate frame {1}
% NOTE: P is not in a homogeneous form (it will be transformed automatically)
P1 = inv(T1) * P

% w.r.t coordinate frame {2}
P2 = inv(T2) * P
```

```matlab
trplot(rotz(30)) % 3D rotation around Z axis
hold on
trplot(eye(3), 'frame', 'o', 'color', 'black') % origin frame
axis equal
```

```matlab
% rotations are not commutative (i.e. order matters)
% so if a frame is rotated for 90 degrees around X and Y
% it is not the same as if a frame is rotated for 90 degress around Y and X
trplot(rotx(90) * roty(90), 'frame', '1', 'color', 'b')
hold on
trplot(roty(90) * rotx(90), 'frame', '2', 'color', 'r')
axis square
```

```matlab
% Euler angles 3D (ZYZ)

R = eul2r(30, 60, 0) % rotation around Z and Y for 30 and 60 deg, accordingly
trplot(eye(3), 'frame', 'o', 'color', 'black')
hold on
trplot(R, 'frame', '1', 'color', 'b')
axis equal
```

```matlab
% an alternative 3D rotation matrix
R = rotz(30) * roty(60)
tr2eul(R, 'deg')
```

```matlab
% roll, pitch, yaw angles (X, Y, and Z)
R = rpy2r(60, 45, 0)

trplot(eye(3), 'frame', 'o', 'color', 'black')
hold on
trplot(R, 'frame', '1', 'color', 'g')
axis equal
```
